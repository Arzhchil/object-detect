```python
import xml.etree.cElementTree as ET

def txt_to_xml(txt_path):
    
    root = ET.Element("annotation")

    size = ET.SubElement(root, "size")
    ET.SubElement(size, "width").text = "some value1"
    ET.SubElement(size, "height").text = "some value1"
    ET.SubElement(size, "depth").text = "some value1"

    obj = ET.SubElement(root, "object")
    ET.SubElement(obj, "name").text = "weapon"

    bndbox = ET.SubElement(obj, "bndbox")
    ET.SubElement(bndbox, "xmin").text = "weapon"
    ET.SubElement(bndbox, "ymin").text = "weapon"
    ET.SubElement(bndbox, "xmax").text = "weapon"
    ET.SubElement(bndbox, "ymax").text = "weapon"

    tree = ET.ElementTree(root)
    tree.write("filename.xml")
```

```python
train_xml_dir = 'data/annotations/train'
test_xml_dir = 'data/annotations/test'
```

```python
train_img_dir = 'data/images/train'
test_img_dir = 'data/images/test'
```

```python
train_xml_files = [f for f in listdir(train_xml_dir) if isfile(join(train_xml_dir, f))]
train_xml_files
```

```python
test_xml_files = [f for f in listdir(test_xml_dir) if isfile(join(test_xml_dir, f))]
test_xml_files
```

```python
len(test_xml_files), len(train_xml_files)
```

```python
for file in test_xml_files[::2]:
    with open(f'{test_xml_dir}/{file}') as f:
        labels = [label.replace('\n', '') for label in f.readlines()]
        width, height = Image.open(f"{test_img_dir}/{file[:-4]}.jpg").size
        root = ET.Element("annotation")
        
        ET.SubElement(root, "folder").text = "test"
        ET.SubElement(root, "filename").text = f"{file[:-4]}.jpg"
        
        size = ET.SubElement(root, "size")
        ET.SubElement(size, "width").text = str(width)
        ET.SubElement(size, "height").text = str(height)
        ET.SubElement(size, "depth").text = "3"

        for string in labels:
            _, x, y, w, h = string.split() 
            
            xmin, ymin, xmax, ymax = float(x) * width - float(w) * width / 2, float(y) * height - float(h) * height / 2, float(x) * width + float(w) * width / 2, float(y) * height + float(h) * height / 2
            
            obj = ET.SubElement(root, "object")
            ET.SubElement(obj, "name").text = "weapon"

            bndbox = ET.SubElement(obj, "bndbox")
            ET.SubElement(bndbox, "xmin").text = str(round(xmin))
            ET.SubElement(bndbox, "ymin").text = str(round(ymin))
            ET.SubElement(bndbox, "xmax").text = str(round(xmax))
            ET.SubElement(bndbox, "ymax").text = str(round(ymax))
            
        tree = ET.ElementTree(root)
        tree.write(f'{test_xml_dir}/{file[:-4]}.xml')
```
